```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# data load
df = pd.read_csv('Instagram data.csv',encoding='latin1')


df.head()
```

```python

df.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Distribution of Impressions")
sns.histplot(df['Impressions'], kde=True)
plt.show()
```

```python

labels = ['From Home', 'From Hashtags', 'From Explore', 'From Other']
values = [df['From Home'].sum(), df['From Hashtags'].sum(), df['From Explore'].sum(), df['From Other'].sum()]


plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Distribution of Impressions by Source')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Likes', y='Comments')
plt.title('Relationship between Likes and Comments')
plt.show()
```

```python
# correlation  heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Instagram Metrics')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# X = input(Impressions), y = target (Likes)
X = df[['Impressions']]
y = df['Likes']

model = LinearRegression()
model.fit(X, y)

print(f"Model Score (Accuracy): {model.score(X, y)}")
```
